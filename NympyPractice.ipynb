```python
def arr_val(arr):
    x=0
    y=0
    z=0
    for i in arr:
        x=x+i
        y=y+i**2
    print(x)
    print(y)
    x=x/len(arr)
    y=m.sqrt(y/len(arr))
    for j in arr:
        z=z+(j-x)**2
    z=m.sqrt(z/len(arr))
    print('Mean, Sq root of mean and std deviation of array is {}, {} and {} respectively'.format(x,y,z))
    
a=([1,2,3,4,5])
b= np.cos(a)**2
arr_val(b)
plt.plot(a,b)
plt.xlabel('Array values : x')
plt.ylabel('Cos(x)^2')
plt.show()
```

```python
arr = np.arange(0 , 2*m.pi,0.10) 
def meanOfArray(parr):
    '''
    Function to calculate the mean of array
    '''
    x = np.sum(parr)/len(parr)
    return x

#print(ar_sum(arr))
def stdOfArray(parr):
    '''
    Function to calculate standard deviation of array
    '''
    y = np.sum((parr-meanOfArray(parr))**2)
    y = m.sqrt(y/len(parr))
    return y
'''
Execution begins from here
'''
print("Mean of array arr[] is {}".format(meanOfArray(arr)))
print("Given Mean of array arr[] is ",str(np.mean(arr)))
print("Standard Deviation of array arr[] is {}".format(stdOfArray(arr)))
print("Given Standard Deviation of array arr[] is ",str(np.std(arr)))

z = np.cos(arr)**2
X = meanOfArray(z)
Y = stdOfArray(z)
print("Mean of array z[] is {}".format(X))
print("Given Mean of array z[] is "+str(np.mean(z)))
print("Standard Deviation of array z[] is {}".format(Y))
print("Given Standard Deviation of array z[] is "+str(np.std(z)))
plt.plot(arr,z)
#plt.plot(Y,X)
plt.xlabel('Array values : x')
plt.ylabel('Cos(x)^2')
plt.show()
```

```python
from functools import reduce
#Use map() filter() reduce() functions 
arr = (1,2,5,8,9)
def productArray(m):
    '''
    returns product of values
    '''
    return m*m
def productValue(m,n):
    '''
    returns product of values take two paramter for reduce()
    '''
    return m*n

result = map(productArray,arr)
print(list(result))
result1 = list(filter(productArray,arr))
print(result1)
result2 = (reduce(productValue,[1,2,4,6]))
print(result2)
```
